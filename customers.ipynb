```python
x=df.iloc[:,2:].values
x
```

```python
scale=MinMaxScaler()
features=scale.fit_transform(x)
features
```

Elbow method

```python
inertia=[]
for i in range(1,16):
  k_means=KMeans(n_clusters=i)
  k_means=k_means.fit(features)
  inertia.append(k_means.inertia_)

```

```python
#elbow graph
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.plot(range(1,16),inertia,'bo-')
plt.xlabel('No.of clusters')
plt.ylabel('Inertia')
```

```python
#elbow visualizer
plt.subplot(1,2,2)
kmeans=KMeans()
visualize=KElbowVisualizer(kmeans,k=(1,16))
visualize.fit(features)
plt.suptitle('visualization')
visualize.poof()
plt.show()
```

```python
#silhoutte score
silhouette_avg=[]
for i in range(2,16):
  kmeans=KMeans(n_clusters=i)
  cluster_labels=kmeans.fit_predict(features)
  silhouette_avg.append(silhouette_score(features,cluster_labels))
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(2,16),silhouette_avg,'bX-')
plt.xlabel('values of k')
plt.ylabel('silhouette score')
plt.title('silhouette analysis for optimal value of k')
plt.show()
```

```python
model=KMeans(n_clusters=3)
model=model.fit(features)
```

```python
y_km=model.predict(features)
centers=model.cluster_centers_
```

```python
df['cluster']=pd.DataFrame(y_km)
df.to_csv("cluster Data",index=False)
```
